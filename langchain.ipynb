```python
###Prompt templates and LLM chain
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompt_template = PromptTemplate(input_variables=['country'],
                                template = "Tell me the capital of this {country}")

prompt_template.format(country = "United States")
```

```python
llm.invoke(prompt_template=prompt_template, text=["India"])
```

```python
from langchain.chains import LLMChain
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template)
print(chain.run("United States"))
```

```python
## Simple Sequential model
```

```python
capital_template = PromptTemplate(input_variables = ['country'],
                               template="Please tell me the capital of {country}")

capital_chain = LLMChain(llm=llm, prompt = capital_template)


famous_template = PromptTemplate(input_variables= ["capital"],
                                template="Suggest me some amazing places to visit in {capital}")

famous_chain = LLMChain(llm=llm, prompt=famous_template)
```

```python
from langchain.chains import SimpleSequentialChain

chain = SimpleSequentialChain(chains=[capital_chain, famous_chain])

chain.run("United States")
```

```python
## Sequential Chain
```

```python
capital_template = PromptTemplate(input_variables = ['country'],
                               template="Please tell me the capital of {country}")

capital_chain = LLMChain(llm=llm, prompt = capital_template, output_key="capital")
```

```python

famous_template = PromptTemplate(input_variables= ["capital"],
                                template="Suggest me some amazing places to visit in {capital}")

famous_chain = LLMChain(llm=llm, prompt=famous_template, output_key ="places")
```

```python
from langchain.chains import SequentialChain
chain = SequentialChain(chains=[capital_chain, famous_chain],
                       input_variables=['country'], 
                       output_variables=['capital', 'places'])
```

```python
chain({'country'=='United States'})
```

```python
### prompt template + LLM + Output parsers
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import ChatPromptTemplate
from langchain.schema import BaseOutputParser
```

```python
class Commaseperatedoutput(BaseOutputParser):
    def parse(self, text:str):
        return text.strip().split(",")
        
```

```python
template = "You are a helpful assistant. When the user gives any input, you should generate 5 words synonyms in a comma seperated list"

human_template = "{text}"
chatprompt = ChatPromptTemplate.from_messages([
    ("system", template),
    ("human", human_template)
    
])
```

```python
chain = chatprompt | chatllm | Commaseperatedoutput()
```

```python
chain.invoke({"text":"intelligent"})
```
